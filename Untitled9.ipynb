```python
class Celular:
    # 1. Metodo __init__ para inicializar atributos (constructor)
    def __init__(self, marca, modelo, precio):
        self.marca = marca
        self.modelo = modelo
        self.precio = precio
        self.encendido = False


    def encender (self):
        self.encendido = True
        print(f"El {self.modelo} de {self.marca} se ha encendido.")

    def mostrar_info(self):
        print(f"Celular: {self.marca} {self.modelo} - Precio: ${self.precio}")

# creacion de objetos(instancias)
celular1 = Celular ("Apple", "iPhone 15", 999)
celular2 = Celular ("Samsung", "Galaxy 524", 899)

# llamadas a las funciones para cada (objeto celular 1 y celular 2
print("celular 1:")
celular1.mostrar_info()
celular1.encender()

print("\ncelular 2:")
celular2.mostrar_info()
celular2.encender()


print("programa terminado")
print("creador de software por Karol Castillo NL = 9")
```
